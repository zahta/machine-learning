#### 64. Print the minimum value and the maximum value for each NumPy scalar type

````python
for dtype in [np.int8, np.int32, np.int64]:
   print("The minimum value of {}: ".format(dtype), np.iinfo(dtype).min)
   print("The maximum value of {}: ".format(dtype),np.iinfo(dtype).max)
for dtype in [np.float32, np.float64]:
   print("The minimum value of {}: ".format(dtype),np.finfo(dtype).min)
   print("The maximum value of {}: ".format(dtype),np.finfo(dtype).max)
````

````python
Ai=[np.int8, np.int32, np.int64]
Af=[np.float32, np.float64]
for dtype in Ai+Af:
    if dtype in Ai:
        print('The minimum value of {}: '.format(dtype),np.iinfo(dtype).min)
        print('The maximum value of {}: '.format(dtype),np.iinfo(dtype).max)
    else:
        print('The minimum value of {}: '.format(dtype),np.finfo(dtype).min)
        print('The minimum value of {}: '.format(dtype),np.finfo(dtype).max)
````

#### 90. Standardize data using Z-Score normalization algorithms

Z-Score formulaï¼
$$Z = \frac{X-\mathrm{mean}(X)}{\mathrm{sd}(X)}$$

````python
# Define the function by formula
def zscore(x, axis = None):
    xmean = x.mean(axis=axis, keepdims=True)
    xstd  = np.std(x, axis=axis, keepdims=True)
    zscore = (x-xmean)/xstd
    return zscore

# Generate random data  
Z = np.random.randint(10, size=(5,5))
print(Z)

zscore(Z)
````

````python
def zscore2(a,axis=None):
    mean=a.mean(axis=axis,keepdims=True)
    std=np.std(a,axis=axis,keepdims=True)
    zscore=(a-mean)/std
    return zscore

Z1=np.random.randint(1,15,size=(7,7))
print(Z1)
zscore2(Z1)
````

#### 91. Standardize data using Min-Max standardized algorithms

Min-Max formulaï¼
$$Y = \frac{Z-\min(Z)}{\max(Z)-\min(Z)}$$

````python
# Define the function by formula
def min_max(x, axis=None):
    min = x.min(axis=axis, keepdims=True)
    max = x.max(axis=axis, keepdims=True)
    result = (x-min)/(max-min)
    return result

# Generate random data
Z = np.random.randint(10, size=(5,5))
print(Z)

min_max(Z)
````

````python
def minmax(a,axis=None):
    min=a.min(axis=axis,keepdims=True)
    max=a.max(axis=axis,keepdims=True)
    y=(a-min)/(max-min)
    return y

Z1=np.random.randint(1,15,size=(7,7))
print(Z1)
minmax(Z1)
````

#### 92. Standardize data using the L2 norm

Formula to get L2 norm:
$$L_2 = \sqrt{x_1^2 + x_2^2 + \ldots + x_i^2}$$

````python
# Define the function by formula
def l2_normalize(v, axis=-1, order=2):
    l2 = np.linalg.norm(v, ord = order, axis=axis, keepdims=True)
    l2[l2==0] = 1
    return v/l2

# Generate random data
Z = np.random.randint(10, size=(5,5))
print(Z)

l2_normalize(Z)
````

````python
def l2_norm(a,axis=-1,order=2):
    l2=np.linalg.norm(a,ord=order,axis=axis,keepdims=True)
    l2[l2==0]=1
    return a/l2

Z1=np.random.randint(1,15,size=(7,7))
print(Z1)
l2_norm(Z1)
````
